```python
import re
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Input text
text = """help me someone please help save me I am in danger there are talking me please stop them get away from me leave me alone call the police reading me I need help please help me I'm trapped please summary help rescue me hurry there after me protect me don't let them hurt me"""


keywords = re.findall(r'\b(help me|save me|i\'m trapped|rescue me|protect me|hurry)\b', text, flags=re.IGNORECASE)

# Check for keywords and send a specific message if found
if keywords:
    # Message to be sent when distress keywords are found
    alert_message = ("URGENT: A distress signal has been detected.\n"
                     "Phrases like 'Help me', 'Rescue me', 'I'm trapped', or 'Protect me' were mentioned.\n"
                     "Immediate action may be required. Please assess the situation promptly.")
    
    def send_alert(message, recipient_emails):
        sender_email = "vivekpawar9274@gmail.com"  # Sender's email
        subject = "Urgent: Distress Alert Detected"
        app_password = "gfccovmasyubdjam"  # App Password for Gmail

        # Create the email
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = ", ".join(recipient_emails)  # Add all recipients to the 'To' field as a comma-separated string
        msg['Subject'] = subject
        msg.attach(MIMEText(message, 'plain'))

        # Send the email to all recipients
        try:
            with smtplib.SMTP('smtp.gmail.com', 587) as server:  # Correct SMTP server for Gmail
                server.starttls()
                server.login(sender_email, app_password)
                server.sendmail(sender_email, recipient_emails, msg.as_string())  # Send to all recipients at once
            print("Email sent successfully to all recipients!")
        except Exception as e:
            print(f"Error sending email: {e}")

    # List of recipients to send the alert to
    recipient_list = ["try.mayursuryavanshi@gmail.com", "21113041@pvgcoet.ac.in","21113044@pvgcoet.ac.in"]

    # Send the alert message to the list of recipients
    send_alert(alert_message, recipient_list)
else:
    print("No distress phrases detected.")
```
